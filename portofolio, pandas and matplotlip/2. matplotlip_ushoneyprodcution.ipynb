```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dfyear = pd.read_csv('df_year_sum.csv')
dfyear
```

## The total weight of US honey production from 1998 to 2012

```python
#Plot

plt.figure(figsize=(8,5))
plt.title('Total production of US honey from 1998 to 2012', fontdict={'fontweight':'bold', 'fontsize': 18})
plt.plot(dfyear.year, dfyear.totalprod)
plt.xlabel('Year')
plt.ylabel('Total honey production (pounds)')
plt.show()
```

```python
labels = dfyear.year
totalprod = dfyear.totalprod
plt.figure(figsize=(8,5), dpi=100)
plt.pie(totalprod, labels = dfyear.year, autopct = '%.2f%%' )
plt.title('Total production of US honey from 1998 to 2012', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.show()
```

## The total weight of US honey production from 1998 to 2012 per state

```python
dfstate = pd.read_csv('df_state_sum.csv')
dfstate
```

```python
#BAR
plt.figure(figsize=(15,12))
plt.xlabel('States')
plt.ylabel('Total honey production (pounds)')
plt.title('Total production of US honey from 1998 to 2012 per state', fontdict={'fontweight':'bold', 'fontsize': 18})

bars = plt.bar(dfstate.state, dfstate.totalprod)

plt.show()
```
